```python
import pandas as pd
from datetime import datetime
from google.colab import files

# Step 1: Upload the file
uploaded = files.upload()  # This will prompt you to upload the file

# Step 2: Load the uploaded file into a pandas DataFrame
file_name = 'CA_daily.xlsx'  # Replace with your uploaded file's name if different
df = pd.read_excel(file_name)

# Filter out rows where Exclusion column is not empty
df_exclude = df[df['Exclusion'].isna() | (df['Exclusion'] == '')]

# REGION: Calculate the average Site Availability (%) based on Region
df_region = df_exclude.groupby('Region')['Site Availability (%)'].mean()

# Calculate the nationwide average by averaging all regions
nationwide_avg = df_exclude['Site Availability (%)'].mean()

# Apply color coding to the nationwide average
def apply_color_coding(availability):
    return "🟢" if availability >= kpi_threshold else "🟡"

# Add the nationwide average as a new row to the result for region-based report
df_region.loc['NATIONWIDE'] = nationwide_avg

# RAN STATE: Calculate the average Site Availability (%) based on RAN State
df_ranstate = df_exclude.groupby('RAN State')['Site Availability (%)'].mean()

# Add the nationwide average for RAN State report
df_ranstate.loc['NATIONWIDE'] = nationwide_avg

# KPI Threshold
kpi_threshold = 99.50

# Get the previous day's date
previous_date = (datetime.now() - pd.Timedelta(days=1)).strftime('%A | %d-%b-%Y')

# Format report for regions
regions_report = "\n".join(
    [f"{apply_color_coding(avail)} {region} = {avail:.2f}%"
     for region, avail in df_region.items()]
)

# Format report for GM states
gm_states_report = "\n".join(
    [f"{apply_color_coding(avail)} {state} = {avail:.2f}%"
     for state, avail in df_ranstate.items()]
)

# Format nationwide CA with color-coding
nationwide_ca_icon = apply_color_coding(nationwide_avg)

# Final report format
report = f"""
*Daily (T-1) Cell Availability with Exclusion*
Report for 📆 {previous_date}

{nationwide_ca_icon} Nationwide = {nationwide_avg:.2f}%

{regions_report}

KPI Threshold: {kpi_threshold}%

*Breaking Down By GM State*

{gm_states_report}

*End Report*
"""

# Print the final report to the console (or save to a file if needed)
print(report)

# Optionally: Save report to a text file with UTF-8 encoding
with open("/content/cell_availability_report.txt", "w", encoding="utf-8") as file:
    file.write(report)

from google.colab import files
files.download("/content/cell_availability_report.txt")
```

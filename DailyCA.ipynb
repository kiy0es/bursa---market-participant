```python
import pandas as pd
from datetime import datetime
from google.colab import files

# Step 1: Upload the file
uploaded = files.upload()  # This will prompt you to upload the file

# Step 2: Get the name of the uploaded file dynamically
file_name = next(iter(uploaded))  # Gets the first uploaded file's name

# Step 3: Load the uploaded file into a pandas DataFrame
df = pd.read_excel(file_name)

# Step 4: Extract the exact date from column "A" (assuming the date is in the first row of column A)
# You might need to adjust the index based on where the date is located in your file
excel_date = pd.to_datetime(df.iloc[0, 0])  # Adjust as per your data's row and column

# Step 5: Use the exact date for the report (no need to calculate previous date)
previous_date = excel_date.strftime('%A | %d-%b-%Y')

# Filter out rows where Exclusion column is not empty
df_exclude = df[df['Exclusion'].isna() | (df['Exclusion'] == '')]

# REGION: Calculate the average Site Availability (%) based on Region
df_region = df_exclude.groupby('Region')['Site Availability (%)'].mean()

# Calculate the nationwide average by averaging all regions
nationwide_avg = df_exclude['Site Availability (%)'].mean()

# Apply color coding to the nationwide average
def apply_color_coding(availability):
    return "🟢" if availability >= kpi_threshold else "🟡"

# RAN STATE: Calculate the average Site Availability (%) based on RAN State
df_ranstate = df_exclude.groupby('RAN State')['Site Availability (%)'].mean()

# KPI Threshold
kpi_threshold = 99.50

# Format report for regions (using title case for region names)
regions_report = "\n".join(
    [f"{apply_color_coding(avail)} {region.title()} = {avail:.2f}%"
     for region, avail in df_region.items()]
)

# Format report for GM states (using title case for state names)
gm_states_report = "\n".join(
    [f"{apply_color_coding(avail)} {state.title()} = {avail:.2f}%"
     for state, avail in df_ranstate.items()]
)

# Format nationwide CA with color-coding
nationwide_ca_icon = apply_color_coding(nationwide_avg)

# Final report format
report = f"""
*Daily (T-1) Cell Availability with Exclusion*
Report for 📆 {previous_date}

{nationwide_ca_icon} Nationwide = {nationwide_avg:.2f}%

{regions_report}

KPI Threshold: {kpi_threshold}%

*Breaking Down By GM State*

{gm_states_report}

*End Report*
"""

# Print the final report to the console (or save to a file if needed)
print(report)

# Overwrite the existing report file with the new content, with a dynamic filename based on the previous date
report_file_path = f"/content/cell_availability_report_{previous_date}.txt"  # Use f-string for dynamic formatting

with open(report_file_path, "w", encoding="utf-8") as file:
    file.write(report)

# Download the overwritten report file
files.download(report_file_path)
```
